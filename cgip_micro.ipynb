```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
# Load the YOLO model from the local .pt file
model = YOLO('/content/yolov8l.pt')

# Load an image
image_path = '/content/vehichles3.webp'
image = cv2.imread(image_path)

# Run inference
results = model(image)

# Extract detections
vehicle_classes = ["car", "truck", "bus", "motorcycle"]
vehicle_count = 0

for r in results:
    for box in r.boxes:
        cls_id = int(box.cls[0])  # Class ID
        label = model.names[cls_id]  # Class name
        if label in vehicle_classes:
            vehicle_count += 1
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(image, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

cv2.putText(image, f"Vehicles: {vehicle_count}", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

cv2_imshow(image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow

# Load the YOLO model from the local .pt file
model = YOLO('/content/yolov8l.pt')

# Load an image
image_path = '/content/vehichles2.png'
image = cv2.imread(image_path)

# Run inference
results = model(image)

# Extract detections
vehicle_classes = ["car", "truck", "bus", "motorcycle"]
vehicle_count = 0

# Counting vehicles
for r in results:
    for box in r.boxes:
        cls_id = int(box.cls[0])  # Class ID
        label = model.names[cls_id]  # Class name
        if label in vehicle_classes:
            vehicle_count += 1
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(image, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Define a threshold for congestion
congestion_threshold = 25  # Example threshold, adjust as needed

# Predict if the traffic is congested
if vehicle_count > congestion_threshold:
    traffic_status = "Congested"
    text_color = (0, 0, 255)  # Red for congested
    # Print warning to the console
    print("WARNING: Traffic is Congested! Vehicle count exceeds threshold.")
else:
    traffic_status = "Not Congested"
    text_color = (0, 255, 0)  # Green for not congested
    # Print a message to the console
    print("Traffic is clear. No congestion detected.")

# Display the vehicle count and traffic status
cv2.putText(image, f"Vehicles: {vehicle_count}", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
cv2.putText(image, f"Traffic: {traffic_status}", (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, text_color, 2)

# Display the image with annotations
cv2_imshow(image)
```
